A simple linear regression model involves a single predictor variable to predict the outcome; and a multiple linear regression model involves two or more independent predictor variables to predict the dependent outcome. Some benefits provides by multiple linear regression includes enhanced predictive power, control for confouding variable, ability to model complex variables, explaining the relationship between variables and control of interaction effects. 

In a simple linear regression model, a continuous predictor variable is able to take any value within a specific range, β1 is used to represent the average affect when one unit of a continuous variable is increased, and x is used to represent the unit changed. An indicator variable takes either a value or 0 or 1, representing two categories such as YES or NO. D is used to represent the indicator variable and controls the groups of coefficient when generating the outcome from different categories. 

y=β0+β1x+ϵ \
where: \
y = dependent variable outcome \
β0 = the intercept \
β1 = the effect of one unit increase in x or y \
x = the continuous independent variable. 

y=β0+β1D+ϵ \
where: \
D = the indicator variable, coded as 0 or 1. 


Introducing an indicator variable longside a continuous variable in a multiple linear regression model allows the model to capture the difference in the relationship between the indicator and continuous variables and the dependent outcome is defined by sorting categories using the indicator variable. 

y=β0+β1x+β2D+ϵ \
where: \
y = outcome \
β0 = intercept \
β1 = slope of x \
x = continuous independent variable \
β2 = the mean difference in y between groups \
D = indicator variable \
where: \
D = 0, y = β0+β1x+ϵ \
D = 1, y = β0+β1x+β2+ϵ

A simple linear regression model only counts the effect of the continuous variable x and assuming there is a uniform relationship for all observation. 

y=β0+β1x+ϵ

A multiple linear regression model with both indicator variable and a continuous variable counts both the effect of the continuous variable x and the indicator variable D, it allows a diffferent intercept for each category and adjusts the model.

Adding an interaction between the indicator variable and continuous variable allows the model to capture a different slope for each category relating to the indicator variable, the relationship between the continuous variable and the dependent outcome can vary in groups, which makes the model more flexible and strength its ability to capture distinct patterns. 

y=β0+β1x+β2D+β3(x⋅D)+ϵ \
where: \
y = outcome \
β0 = intercept \
β1 = slope of x \
x = continuous variable \
β2 = mean shift when D = 1 \
D = indicator variable \
β3 = coefficient for the interaction term

When a multiple linear regression model involves only indicator variables derived from a non-binary categorical variable, it divides the euqation into multiple terms of indicator variable and model the mean differences between multiple categories of categorical predictor and serves to estimate the average of each category. 

y=β0+β1D1+β2D2+⋯+βk−1Dk−1+ϵ \

y = dependent outcome \

```python
import numpy as np # imports numpy
from sklearn.model_selection import train_test_split # imports train_test_split form sklearn

# pokeman.shape[0] returns the numbers of row
# * 0.5 returrns the size of half the dataset
fifty_fifty_split_size = int(pokeaman.shape[0]*0.5) # calculates the split size of a 50-50 division for the dataset

# Replace "NaN" (in the "Type 2" column with "None")
pokeaman.fillna('None', inplace=True) # replaces miising values 

np.random.seed(130) # sest random seeds and ensures reproducibility
pokeaman_train,pokeaman_test = \
  train_test_split(pokeaman, train_size=fifty_fifty_split_size)# splits pokeaman into poekeaman_train and pokeaman_test
pokeaman_train
```

```python
model_spec3 = smf.ols(formula='HP ~ Attack + Defense', data=pokeaman_train) # defines a OLS model
# HP is used as the dependent variable
# Attack and Defense are used as the independent variables
# illustrates how the defense and attack influence HP

model3_fit = model_spec3.fit() # fits the model to pokeaman_train, estimates the coefficients for independent variables

model3_fit.summary() # detailed summary for linear regression model
```

The result of this cell generates a mutiplie linear regression model with two independent variables 'Attack' and 'Defense', and gives an estimation for the dependent variable 'HP'.
```python
yhat_model3 = model3_fit.predict(pokeaman_test) # use the predict method to generate predictions for HP
y = pokeaman_test.HP # assigns HP to be the y value
print("'In sample' R-squared:    ", model3_fit.rsquared) # "in-sample" R-squared is retrieved directly from the model summary using model3_fit.rsquared
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model3)[0,1]**2) # calculate 'out of sample' R-squared value using the correlation ceofficient between the actual and predicted values. 
```

```python
import pandas as pd

# Load the Pokémon dataset
url = "https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv"
pokemon = pd.read_csv(url)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
import statsmodels.formula.api as smf

# Initialize lists to store in-sample and out-of-sample R-squared values
in_sample_r_squared = []
out_of_sample_r_squared = []

# Define the model formula
model_formula = 'HP ~ Attack + Speed + Q("Sp. Def") + Q("Sp. Atk") + I(Q("Type 1")=="Normal") + I(Q("Type 1")=="Water") + I(Generation==2) + I(Generation==5)'

# Run the loop for multiple iterations
for _ in range(50):  # Run 50 iterations for robust sampling
    # Split the data randomly into training and test sets
    pokeaman_train, pokeaman_test = train_test_split(pokemon, train_size=0.7)
    
    # Create and fit the model on the training set
    model_spec = smf.ols(formula=model_formula, data=pokeaman_train)
    model_fit = model_spec.fit()
    
    # Calculate "in-sample" R-squared
    in_sample_r_squared.append(model_fit.rsquared)
    
    # Calculate "out-of-sample" R-squared
    yhat_test = model_fit.predict(pokeaman_test)
    y_test = pokeaman_test.HP
    out_of_sample_r_squared.append(np.corrcoef(y_test, yhat_test)[0,1]**2)

# Visualize the in-sample and out-of-sample R-squared values
plt.figure(figsize=(10, 6))
plt.plot(in_sample_r_squared, label="In-sample R-squared", marker='o', linestyle='-')
plt.plot(out_of_sample_r_squared, label="Out-of-sample R-squared", marker='x', linestyle='--')
plt.xlabel("Iteration")
plt.ylabel("R-squared")
plt.title("In-sample vs Out-of-sample R-squared Across Different Data Splits")
plt.legend()
plt.show()
```

Variability in 'in-sample' R-squared values tells how the model's sensitivity to changes in the training data and suggests that the model captures stable partterns within the training data. Variability in'out-of-sample' R-squared values reflects the model's ability to generalize. In this plot, it suggests the model is not generalizes well acroos different test sets.
